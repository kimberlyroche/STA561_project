```python
import pandas
from pandas.core.series import Series

data = pandas.DataFrame()

movies = ['Movie 1', 'Movie 2']
genres = ['Drama', 'Action']

for movie in movies:
    for genre in genres:
        data = data.append(Series({'movie': movie, 'genre':genre}), ignore_index=True)

movies = ['Movie 3', 'Movie 4']
genres = ['Drama', 'Comedy']

for movie in movies:
    for genre in genres:
        data = data.append(Series({'movie': movie, 'genre':genre}), ignore_index=True)
        
data['count'] = 1
```
